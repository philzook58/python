```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import cm

```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

b= np.sqrt(3)/2
a = 3/2
def energy(kx, ky):
    return np.sqrt(1+ 4 * np.cos( b * ky) * np.cos(a * kx) + 4 * np.cos( b * ky) ** 2)

# Make data.
X = np.arange(- np.pi, np.pi, 0.01)
Y = np.arange(-np.pi, np.pi, 0.01)
X, Y = np.meshgrid(X, Y)
E = energy(X,Y)


# Plot the surface.
surf = ax.plot_surface(X, Y, E, cmap=cm.coolwarm, linewidth=0 )
surf = ax.plot_surface(X, Y, -1 * E,  linewidth=0 )
```
