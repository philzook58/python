```python
def schurcomplement(M,n,m):
        A = M[:n,:m]
        B = M[n:,:m]
        C = M[:n,m:]
        D = M[n:,m:]
        return A - np.dot(np.dot(B,np.linalg.inv(D)),C)
```

```python
schurcomplement(np.array([[1,2],[3,4]]),1,1)
```

Some schur complement experiments. Doing iterative schur complement builds up an $i\epsilon$ into a bigger imaginary component

```python
t=2.0
e=0.01j
Heff = np.array([[e]])
for i in range(1000):
    H = np.array([[e,t],
                  [t,Heff[0,0]]])
    Heff = schurcomplement(H,1,1)
    print Heff

    
```

```python
def H(E):
    return (e-E - np.sqrt((e-E)**2-4* t**2))/2.

es = np.linspace(-6,6,100)
plt.plot(es,np.real(H(es)))
plt.plot(es,np.imag(H(es)))
```
