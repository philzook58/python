```python
import numpy as np
```

```python
N = 200
p1 = np.random.rand(1,N,2)
p2 = np.random.rand(N,1,2)
p2[:,0,0] += 4
```

```python
dr = np.linalg.norm(np.repeat(p2, N, axis=1) - np.repeat(p1, N, axis=0), axis=2)
G = np.log(dr)
```

Here's a question: 2D helmholtz equation, does rank depend on etendue?

```python
import scipy.special
```

```python

def H0(k, r):
    return scipy.special.hankel1(0,k * r)

def H0norm(k, x):
    return scipy.special.hankel1(0,k * np.linalg.norm(x,axis=2))
```

```python
x = np.linspace(-20,20,100)
plt.plot(x,np.abs(H0(1,x)))
plt.plot(x,np.real(H0(1,x)))
plt.plot(x,np.imag(H0(1,x)))
plt.plot(x,np.sqrt(2/np.pi)/np.sqrt(x))
```

```python
class Quadtree():
    def __init__(self, N):
        self.vecs = []
        for i in range(N):
            self.vecs.append(np.array.zeros(4**i))
    def val(i, j, level):
        return self.vecs[i,j]
    def __getitem__(self, index):
        x = index[0]
        y = index[1]
        level = index[2]
        ''' #To do later, check for slicing
        if type(x) == slice:
            x.start
            x.stop
            x.step
            '''
        newindex =  x + y * 2 ** level 
            
        return self.vecs[level][newindex]
            
def quadindex(i, j, level):
    return x + y * 2 ** level + quadsize(level-1) 

# return the total size of a quadtreevector of total length 2^N on a side. or with 4^N boxes
#at lowest level
# 4^N + 4^(N-1)+ ... 1 =  ( 1 - 4^(N+1)) / (1-4)
def quadsize(N):
    return (1 - 4 ** (N-1)) / (1-4)

def quadchildren(i, j, level):
    return [(2* i, 2 * j,level+1),(2 * i+ 1, 2 * j,level+1),(2 * i, 2 * j+1,level+1),(2*i+ 1, 2* j + 1,level+1) ]
    
def quadparent(i,j, level):
    return (i//2,j//2,level-1)
    
def quadneighbor(i, j, level):
    return [(i+ 1, j,level),(i, j+1,level),(i+1 ,j + 1,level),(i- 1, j,level),(i, j-1,level),(i-1 ,j - 1,level) ]
    
```
