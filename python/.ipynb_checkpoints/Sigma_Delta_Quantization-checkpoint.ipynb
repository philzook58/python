```python
import numpy as np
import scipy as sp
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
T = 10.0
N = 10000
t = np.linspace(0,T,N)

def plotspec(y):
    plt.plot(np.fft.fftfreq(N,T/N)[:100] , np.log(0.001+np.abs(sp.fft(y)[:100])))

def plotfullspec(y):
    plt.plot(np.fft.fftfreq(N,T/N) , np.log(0.001+np.abs(sp.fft(y))))
    
def bitquantize(y):
    return np.sign(y)  
```

```python
y = np.cos(2*np.pi*t)
plt.plot(t,y)
plt.figure()
plotspec(y)
```

```python
quant = bitquantize(y)
plt.plot(t,quant)
plt.figure()
plotspec(quant)
```

Dithering

```python
quant = bitquantize(y) + np.random.normal(0,2.0, N)
plt.plot(t,quant)
plt.figure()
plotspec(quant)
```

```python
quant = bitquantize(y + np.random.normal(0,1.0, N))
plt.plot(t,quant)
plt.figure()
plotspec(quant)
```

```python
quant = np.zeros(N)
quant[0] = bitquantize(y[0])
integ = 0.0
for i in range(1,N):
    integ += (y[i-1] - quant[i-1]) * T/N
    quant[i] =  bitquantize(integ)
plt.plot(t,quant)
plt.figure()
plt.plot(t[:500],quant[:500])
plt.figure()
plotspec(quant)
plt.figure()
plotfullspec(quant)
```

Noise is significantly depressed at low freqeuency

```python
low = np.zeros(N)
low[:100] = 1.0
low[N-101:N] = 1.0
plt.plot(sp.ifft(sp.fft(quant)*low))
plt.figure()
plt.plot(np.fft.ifft(sp.fft(quant)*low)-y)
```

Simple Decimation

```python

lowpassed = (quant[::4]+quant[1::4] +quant[2::4]+quant[3::4])/4
plt.plot(lowpassed)
plt.figure()
plt.plot(lowpassed-y[::4])
```

```python
decimationfactor = 13 
lowpassed = sp.signal.decimate(quant, decimationfactor)
plt.plot(lowpassed)
plt.figure()
plt.plot(lowpassed-y[::decimationfactor])
```

Weird. Clearly has a delay.
