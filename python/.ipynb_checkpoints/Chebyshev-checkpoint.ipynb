```python
import numpy as np
```

```python
def chebpoints(n):
    return np.cos(np.arange(n) * np.pi / (n-1))
```

This form is silly. Exponential work. The memoized form would be acceptable

```python
def chebeval(x,n):
    if n == 0:
        return 1
    if n == 1:
        return x
    return 2 * x * chebeval(x, n-1) -  chebeval(x, n - 2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-1,1, 500)
plt.plot(x, chebeval(x,20))
```

```python
from scipy.fftpack import dct
def chebpoints2(n):
    return np.cos(np.arange(1,n-1) * np.pi / (n-1))
def chebsample(f,n):
    x = chebpoints2(n)
    vals = f(x)
    a = dct(vals)
    return a

def intchebfactor(n):
    if n == 1:
        return 0
    else:
        ((-1)**n + 1)/(1 - n*n)
```
