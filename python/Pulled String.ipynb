```python
from scipy.linalg import toeplitz
N = 20
firstcol = np.zeros(N)
firstcol[0] = -2
firstcol[1] = 1
firstrow = np.zeros(N)
firstrow[0] = -2
firstrow[1] = 1
K = toeplitz(firstcol, firstrow)
K = -1 * K
G = np.linalg.inv(K)
plt.plot(G[:,:])
#plt.ylim([-1,1])
```

```python
eigsK = np.linalg.eigvalsh(K)
exacteigs = (np.pi * np.arange(1,21)  /22)**2
plt.plot(eigsK)
plt.plot(exacteigs)
```
